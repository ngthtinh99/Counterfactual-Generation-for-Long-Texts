Convert the `results_raw.csv` file into `my_results.csv` (or `results.csv`) with a proper format that could be evaluated by CEval.

```python
import pandas as pd

# Read the CSV file
df_output = pd.read_csv('results_raw.csv')

# Add a new column 'gen_text_2' as a copy of 'gen_text'
df_output['gen_text_2'] = df_output['gen_text']

# Save the modified DataFrame back to the same CSV file
df_output.to_csv('my_results.csv', index=False)

print("Column 'gen_text_2' added as a copy of 'gen_text' and file 'my_results.csv' updated successfully.")
```
